# É importante fazer as comparaçãoes entre dataframes diferentes com mesmos dados

```python
notas = pd.read_csv('ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby(['movieId']).mean()['rating']
nota_media_por_filme.head()
```

```python
ax = sns.histplot(nota_media_por_filme.values)
ax.set(xlabel='Nota Média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no Movielens 100k')
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos
```

```python
ax = sns.histplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel='Nota Média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no Movielens 100k com 10 votos ou mais')
```

# Comparação entre Toy Story e os demais filmes

```python
filmes = pd.read_csv('movies.csv')
filmes.query('movieId == 1')
```

```python
notas1 = notas.query('movieId == 1')
notas1.head()
```

```python
ax = sns.histplot(notas1.rating)
ax.set(xlabel='Nota', ylabel='Densidade')
ax.set_title('Distribuição das notas para Toy Story')
```

```python
ax = sns.boxplot(x=notas1.rating)
ax.set(xlabel='Nota')
ax.set_title('Distribuição de notas de Toy Story')
```

```python
notas1.rating.mean()
```

```python
zconfint(notas1.rating)
```

H0: media_notas1 = 3.4320503405352603

```python
from statsmodels.stats.weightstats import ztest

_, p_valor = ztest(notas1.rating, value=3.4320503405352603)
```

```python
p_valor <= 0.05
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    _ , p_value = ztest(temp[0:i], value=3.4320503405352603)

    return (i, media, p_value)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:, 0], valores[:, 1])
plt.plot(valores[:, 0], valores[:, 2])
plt.hlines(y=0.05, xmin=2, xmax=len(temp), colors='red')
```
